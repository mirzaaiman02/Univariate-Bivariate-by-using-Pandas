```python
## import pandas and load imdb_rating.xlsx data
import pandas as pd
xls=pd.ExcelFile('imdb_rating.xlsx')
df = xls.parse('imdb_rating')
print(type(df))
```

```python
# Define the columns to cleaning
columns_to_clean = ['Budget', 'Gross USA', 'Cumulative worlwide Gross']

# Remove $, commas, "estimated", and parentheses
for col in columns_to_clean:
    df[col] = df[col].replace({'\$': '', ',': '', 'estimated': '', '\(': '', '\)': ''}, regex=True)

# Drop NA values from these columns which have values NaN
df.dropna(subset=columns_to_clean, inplace=True)

# Convert columns to numeric
df[columns_to_clean] = df[columns_to_clean].apply(pd.to_numeric, errors='coerce')

# Drop any rows that failed conversion (still contain non-numeric values)
df.dropna(subset=columns_to_clean, inplace=True)

```

```python
## to see top 10 from above the result after done process of cleaning
df.head(10)
```

```python
## Histogram and KDE Plot for 'IMDB rating'
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
df.hist(column=['IMDB rating'], grid=False, edgecolor='black')
plt.title('Histogram of IMDB Ratings')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.show()

import seaborn as sns

# KDE Plot
sns.kdeplot(df['IMDB rating'])
plt.title('KDE Plot of IMDB Ratings')
plt.xlabel('IMDB Rating')
plt.ylabel('Density')
plt.show()
```

This is the findings based on histogram and kdeplot for attributes IMDB rating:

- The histogram and KDE plot for IMDB ratings indicate that most ratings cluster around an average value, creating a bell-shaped curve.
- So, this means that the majority of movies have ratings near the middle, with fewer movies receiving very low or very high ratings.
- There's a slight tendency towards higher ratings, which might imply that the dataset contains more highly rated movies,
  or that people are more likely to rate movies they enjoyed.

Using pandas calculate the T-Test then discuss about the findings of the significance all of the linear regression models obtained from the previous question.

```python
import pandas as pd
import statsmodels.api as sm
```

```python
xls = pd.ExcelFile("imdb_rating.xlsx")
df = xls.parse("imdb_rating")
```

```python
# Selecting only the relevant columns
selected_columns = ['Year released', 'IMDB rating', 'Number of reviewers']
df_selected = df[selected_columns]

# Display the first few rows of the selected DataFrame
(df_selected.head())
```

```python
import statsmodels.api as sm

# Function to perform linear regression and T-test
def linear_regression_ttest(x, y):
    X = sm.add_constant(x)  # Adds a constant term to the predictor
    model = sm.OLS(y, X).fit()  # Fit the ordinary least squares model
    t_test = model.t_test([0, 1])  # T-test for the slope
    return model, t_test

# IMDB Rating vs Number of Reviewers
x1 = df_selected['Number of reviewers']
y1 = df_selected['IMDB rating']
model1, ttest1 = linear_regression_ttest(x1, y1)
print("IMDB Rating vs Number of Reviewers T-Test:\n", ttest1)

# IMDB Rating vs Year Released
x2 = df_selected['Year released']
y2 = df_selected['IMDB rating']
model2, ttest2 = linear_regression_ttest(x2, y2)
print("IMDB Rating vs Year Released T-Test:\n", ttest2)

# Number of Reviewers vs Year Released
x3 = df_selected['Year released']
y3 = df_selected['Number of reviewers']
model3, ttest3 = linear_regression_ttest(x3, y3)
print("Number of Reviewers vs Year Released T-Test:\n", ttest3)
```
